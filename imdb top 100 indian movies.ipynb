IMDB TOP 100 Indian movies

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
names=soup.find_all('td',class_='titleColumn')
names
```

```python
movies=[]
for i in names:
    for j in i.find_all('a'):
        movies.append(j.text)
movies
```

```python
ratings=soup.find_all('td',class_="ratingColumn imdbRating")
ratings
```

```python
rating=[]
for i in ratings:
    rating.append(i.text.replace('\n',''))
rating
    
```

```python
year=soup.find_all('span', class_="secondaryInfo")
year
```

```python
release_year=[]
for i in year:
    release_year.append(i.text)
release_year
```

```python
print(len(movies),len(rating),len(release_year))
```

```python
import pandas as pd
IMDB_Indian100=pd.DataFrame({})
IMDB_Indian100['Movie_name']=movies[:100]
IMDB_Indian100['rating']=ratings[:100]
IMDB_Indian100['year_of_release']=release_year[:100]




```

```python
IMDB_Indian100
```

```python
IMDB_Indian100.to_csv('imdb top100 indian movies.csv')
```
