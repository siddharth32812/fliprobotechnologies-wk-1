```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.dineout.co.in/delhi-restaurants")
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
first_title=soup.find('div',class_="restnt-info cursor" )
first_title
```

```python
first_title.text
```

```python
loc=soup.find('div',class_="restnt-loc ellipsis")
loc
```

```python
loc.text
```

```python
sta=soup.find('span',class_="double-line-ellipsis")
sta
```

```python
sta.text
```

```python
titles=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)

titles
```

```python
locations=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    locations.append(i.text)
locations
```

```python
price=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text)
price
```

```python
images=[]
for i in soup.find_all("img",class_="no-img"):
    images.append(i['data-src'])
    
images
```

```python
print(len(titles),len(locations),len(price),len(images))
```

```python
import pandas as pd
df=pd.DataFrame({'Titles':titles,'Location':locations,'Price':price,'Images_url':images})
df
```

```python
df.to_csv('Dineout.csv')
```
